```python
import pandas as pd 
import numpy as np 
import tensorflow as tf 
from sklearn import model_selection
```

```python
data=pd.read_excel("ENB2012_data.xlsx",0)
data.head()
```

```python
data.shape
```

```python
data=data.drop(columns="Y2")
data.head()
```

```python
y=data[["Y1"]]
y
```

```python
x=data.drop(columns="Y1")
x
```

```python
X=x.to_numpy()
X
```

```python
Y=y.to_numpy()
Y
```

# model building 

```python
trainx,trainy,testx,testy=model_selection.train_test_split(X,Y,train_size=.70,random_state=3)
```

```python
epoch=10
batch_size=70
```

```python
model1=tf.keras.models.Sequential([
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512,activation=tf.nn.relu),
    tf.keras.layers.Dropout(.2),
    tf.keras.layers.Dense(256,activation=tf.nn.relu),
    tf.keras.layers.Dropout(.2),
    tf.keras.layers.Dense(128,activation=tf.nn.log_softmax),
    tf.keras.layers.Dropout(.2),
    tf.keras.layers.Dense(1,activation='linear')])
```

```python
optimiser=tf.keras.optimizers.Adam()
model1.compile(optimizer=optimiser,loss=tf.keras.losses.huber,metrics=tf.keras.metrics.mean_absolute_error)
```

```python
model1.fit(trainx,testx,batch_size=70,epochs=5)
```

```python
model1.evaluate(trainy, testy)
```
